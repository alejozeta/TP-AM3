



# Senales de Entrada

Sine wave signal

```python
import numpy as np
from scipy.io.wavfile import write

# Parameters
samplerate = 44100  # Sample rate in Hz
duration = 5  # Duration in seconds
frequency = 440  # Frequency of the sine wave in Hz

# Generate time points
t = np.linspace(0, duration, int(samplerate * duration), endpoint=False)

# Generate sine wave
sine_wave = 0.5 * np.sin(2 * np.pi * frequency * t)

# Save to a .wav file
write('sine_wave.wav', samplerate, sine_wave.astype(np.float32))
```

White Noise Signal


```python
# Generate white noise
white_noise = np.random.normal(0, 1, int(samplerate * duration))

# Save to a .wav file
write('white_noise.wav', samplerate, white_noise.astype(np.float32))
```

Chirp Signal (La que uso Ale en la tuto.)

```python
from scipy.signal import chirp

# Generate chirp signal (sweep from 20 Hz to 20 kHz)
sweep = chirp(t, f0=20, f1=20000, t1=duration, method='linear')

# Save to a .wav file
write('sweep.wav', samplerate, sweep.astype(np.float32))
```

Square Signal

```python
from scipy import signal

# Generate square wave
square_wave = 0.5 * signal.square(2 * np.pi * frequency * t)

# Save to a .wav file
write('square_wave.wav', samplerate, square_wave.astype(np.float32))
```

Impulse Signal

```python
# Generate impulse (a single sample with value 1, rest 0)
impulse = np.zeros(int(samplerate * duration))
impulse[0] = 1  # Set first sample to 1

# Save to a .wav file
write('impulse.wav', samplerate, impulse.astype(np.float32))
```

Silence Signal

```python
# Generate silence
silence = np.zeros(int(samplerate * duration))

# Save to a .wav file
write('silence.wav', samplerate, silence.astype(np.float32))
```

Saw Tooth Wave (Back to back square signal)

```python
# Generate sawtooth wave
sawtooth_wave = 0.5 * signal.sawtooth(2 * np.pi * frequency * t)

# Save to a .wav file
write('sawtooth_wave.wav', samplerate, sawtooth_wave.astype(np.float32))
```

```python
import numpy as np
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Funcion para abrir archivo .wav
def open_wav(file_path):
    sample_rate, data = wavfile.read(file_path)
    return sample_rate, data

# Calcular magnitud y fase de la FFT
def apply_fft(data, sample_rate):
    # Perform FFT
    N = len(data)
    yf = fft(data)

    # obtener frecuencias
    xf = fftfreq(N, 1 / sample_rate)

    # Calcular magnitud y fase
    magnitude = np.abs(yf)
    phase = np.angle(yf)

    return xf[:N // 2], magnitude[:N // 2], phase[:N // 2] # devolver solo frecuencias positivas

# Dibuja el diagrama de Bode
def plot_bode(frequencies, magnitude, phase):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

    # dibujar magnitud
    ax1.plot(frequencies, 20 * np.log10(magnitude))
    ax1.set_title('Bode Diagram')
    ax1.set_ylabel('Magnitude (dB)')
    ax1.set_xscale('log')
    ax1.grid()

    # dibujar fase
    ax2.plot(frequencies, phase * 180 / np.pi)
    ax2.set_ylabel('Phase (degrees)')
    ax2.set_xlabel('Frequency (Hz)')
    ax2.set_xscale('log')
    ax2.grid()

    plt.tight_layout()
    plt.show()

# Funcion que calcula y dibuja el diagrama de Bode de un archivo .wav
def bode_diagram_for_wav(file_path):
    sample_rate, data = open_wav(file_path)

    # Si el archivo tiene mas de un canal, se toma solo el primer canal
    if data.ndim > 1:
        data = data[:, 0]

    frequencies, magnitude, phase = apply_fft(data, sample_rate)
    plot_bode(frequencies, magnitude, phase)

wav_file_path = 'white_noise.wav'
bode_diagram_for_wav(wav_file_path)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
from pydub import AudioSegment

# Function to open an MP3 file and convert to sample rate and data
def open_mp3(file_path):
    audio = AudioSegment.from_mp3(file_path)
    sample_rate = audio.frame_rate
    data = np.array(audio.get_array_of_samples())
    return sample_rate, data

# Calculate magnitude and phase of FFT
def apply_fft(data, sample_rate):
    N = len(data)
    yf = fft(data)
    xf = fftfreq(N, 1 / sample_rate)
    magnitude = np.abs(yf)
    phase = np.angle(yf)
    return xf[:N // 2], magnitude[:N // 2], phase[:N // 2]  # Return positive frequencies only

# Plot Bode diagram
def plot_bode(frequencies, magnitude, phase, title):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

    # Plot magnitude
    ax1.plot(frequencies, 20 * np.log10(magnitude))
    ax1.set_title(f'Bode Diagram - {title}')
    ax1.set_ylabel('Magnitude (dB)')
    ax1.set_xscale('log')
    ax1.grid()

    # Plot phase
    ax2.plot(frequencies, phase * 180 / np.pi)
    ax2.set_ylabel('Phase (degrees)')
    ax2.set_xlabel('Frequency (Hz)')
    ax2.set_xscale('log')
    ax2.grid()

    plt.tight_layout()
    plt.show()

# Function to compute and plot Bode diagram for an MP3 file
def bode_diagram_for_mp3(file_path, title):
    sample_rate, data = open_mp3(file_path)

    # If the audio has more than one channel (stereo), use only the first channel
    if data.ndim > 1:
        data = data[:, 0]

    frequencies, magnitude, phase = apply_fft(data, sample_rate)
    plot_bode(frequencies, magnitude, phase, title)

# File paths for each signal
file_paths = {
    'Impulse': 'impulse_output.mp3',
    'Sawtooth': 'sawtooth_wave_output.mp3',
    'Silence': 'silence_output.mp3',
    'Sine': 'sine_wave_output.mp3',
    'Square': 'square_wave_output.mp3',
    'Sweep': 'sweep_output.mp3',
    'White Noise': 'white_noise_output.mp3'
}

# Generate Bode diagrams for each file
for title, file_path in file_paths.items():
    bode_diagram_for_mp3(file_path, title)
```
